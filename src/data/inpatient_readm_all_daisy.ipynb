```python
import pandas as pd
import os
import numpy
```

```python
DATA_PATH = "/home/bhatti/dataset/VCHAMPS"
```

```python
# dataset
# useful column: Internalpatientid, Age at admission, Admission date, 
#                First listed discharge diagnosis icd10 subcategory, Second listed discharge diagnosis icd10 subcategory,
#                Died during admission, Outpatientreferralflag
dataset_path = 'inpatient_admissions_train.csv'
load_path = os.path.join(DATA_PATH, dataset_path)
in_ad = pd.read_csv(load_path).iloc[:,1:]
in_ad['Hospital readmission'] = 0
in_ad.rename(columns = {'Died during admission':'Mortality'})
in_ad.head()
```

```python
for i in range(len(in_ad['Internalpatientid'])):
    pid = in_ad['Internalpatientid'].iloc[i]
    if len(in_ad[in_ad['Internalpatientid'] == pid]) > 1:
        early_time = min(in_ad[in_ad['Internalpatientid'] == pid]['Admission date'])
        if in_ad['Admission date'].iloc[i] == early_time:
            in_ad['Hospital readmission'].iloc[i] = 0
        else:
            in_ad['Hospital readmission'].iloc[i] = 1
    else:
        in_ad['Hospital readmission'].iloc[i] = 0

```
