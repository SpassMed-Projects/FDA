```python
import pandas as pd


import numpy as np
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline

from datetime import datetime
```

```python
path = '/home/hassan/lily/MLA/FDA/outpatient_mortality_quality.csv'
outpatient = pd.read_csv(path).iloc[:,1:]
outpatient.head()
```

```python
path = '/data/public/MLA/VCHAMPS-Quality/procedures_qual.csv'
procedures = pd.read_csv(path).iloc[:,1:]
procedures
```

```python
procedures = procedures.drop_duplicates(subset = ['Internalpatientid', 'Procedure date','Procedure code description'],keep = 'first')
```

```python
outpatient_date = outpatient[['Internalpatientid','last_visit_date']]
outpatient_date = outpatient_date.drop_duplicates(keep = 'first')
outpatient_date
```

```python
# Filter procedure date that's before discharge date
outpatient_pro = procedures.merge(outpatient_date, how = 'left', on = 'Internalpatientid')
outpatient_pro2 = outpatient_pro.loc[outpatient_pro['Procedure date'] <= outpatient_pro['last_visit_date']]
outpatient_pro2 
```

```python
# Calculate total procedures before last admission for each patient
outpatient_pro2 ['total_procedure'] = outpatient_pro2['Procedure code description'].groupby(outpatient_pro2['Internalpatientid']).transform('count')
```

```python
# Calculate total procedures that's surgery before last admission for each patient
surgery_pro = outpatient_pro2.loc[outpatient_pro2['Procedure code description'].str.contains('SURG')]
surgery_pro['num_surgery_pro'] = surgery_pro['Procedure code'].groupby(surgery_pro['Internalpatientid']).transform('count')
```

```python
procedures['Internalpatientid'].nunique()
```

```python
columns = ['Internalpatientid', 'total_procedure']
df = outpatient_pro2[columns]
df = df.groupby('Internalpatientid').mean().reset_index(drop = False)
df = df.merge(surgery_pro[['Internalpatientid','num_surgery_pro']], how = 'left',on = 'Internalpatientid')
df['num_surgery_pro'] = df['num_surgery_pro'].fillna(0)
df = df.groupby('Internalpatientid').mean().reset_index(drop = False)
df
```
